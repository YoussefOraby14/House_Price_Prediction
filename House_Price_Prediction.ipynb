Import dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset


```python
house_price_dataset = fetch_california_housing(as_frame=True)
```

```python
print(house_price_dataset)
```

```python
house_price_frames = pd.DataFrame(house_price_dataset.data)
```

```python
house_price_frames.head()
```

```python
#add a column target (price) to the datasetframe
house_price_frames['Price'] = house_price_dataset.target
house_price_frames.head()
```

```python
#number of rows and features
house_price_frames.shape
```

```python
#checking for missing values
house_price_frames.isnull().sum()
```

```python
#statistical measures of the dataset
house_price_frames.describe()
```

Understanding the correlation between various features in the dataset


1- Postive Correlation.
2- Negative Correlation.


```python
correlation = house_price_frames.corr()
```

```python
#constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
#
```

Splitting the Data and Target

```python
X = house_price_frames.drop(['Price'], axis=1)
Y = house_price_frames['Price']
```

```python
print(X)
print(Y)
```

Splitting the data into train and test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor


```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with x_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on the taining data

```python
#accuracy for prediction on trainig data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
#R squared error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared error : ", error_score)

#Mean Abs error
error_score = metrics.mean_absolute_error(Y_train, training_data_prediction)
print("Mean Abs error : ", error_score)
#
```

Visualizing the actual Prices and Predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

Prediction of the Test Data

```python
#accuracy for prediction on testing data
testing_data_prediction = model.predict(X_test)
print(testing_data_prediction)
```

```python
#R squared error
error_score = metrics.r2_score(Y_test, testing_data_prediction)
print("R squared error : ", error_score)

#Mean Abs error
error_score = metrics.mean_absolute_error(Y_test, testing_data_prediction)
print("Mean Abs error : ", error_score)
#
```

```python
plt.scatter(Y_test, testing_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
